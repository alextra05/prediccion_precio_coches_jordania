```python
from sklearn.decomposition import PCA
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 1. Cargar datos
X, y, cat_features, num_features = load_and_prepare_data_car_dataset(df_train)

# 2. Preprocesamiento completo
num_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

cat_transformer = Pipeline([
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("encoder", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

preprocessor = ColumnTransformer([
    ("num", num_transformer, num_features),
    ("cat", cat_transformer, cat_features)
])

# 3. Aplicar preprocesamiento + PCA
pipeline_pca = Pipeline([
    ("preprocessor", preprocessor),
    ("pca", PCA(n_components=2))
])

X_reduced = pipeline_pca.fit_transform(X)

# 4. Entrenar clasificador en 2D
X_train_2D, X_test_2D, y_train, y_test = train_test_split(X_reduced, y, test_size=0.2, stratify=y, random_state=42)
clf = LogisticRegression()
clf.fit(X_train_2D, y_train)

# 5. Visualizar frontera de decisión
def plot_decision_boundary(X, y, model, title):
    h = 0.02
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.figure(figsize=(8, 6))
    plt.contourf(xx, yy, Z, alpha=0.3, cmap="coolwarm")
    sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y, palette="coolwarm", edgecolor="k", alpha=0.8)
    plt.title(title)
    plt.xlabel("PCA Component 1")
    plt.ylabel("PCA Component 2")
    plt.legend(title="Price > 20K")
    plt.grid(True)
    plt.tight_layout()
    plt.show()

plot_decision_boundary(X_test_2D, y_test, clf, "Frontera de decisión tras reducción con PCA")
```

```python

Frontera lineal ineficaz: La frontera de decisión generada por LogisticRegression en el espacio reducido (2 componentes principales) no separa adecuadamente las clases. Esto indica que las clases no son linealmente separables en este espacio.
Pérdida de información: La reducción de dimensionalidad con PCA, aunque útil para visualización, puede haber eliminado información relevante para la clasificación.
Solapamiento de clases: Las clases están fuertemente solapadas en el espacio reducido, lo que sugiere que las componentes principales no capturan una separación clara entre ellas.
Concentración en un eje: La mayoría de los datos se concentran en torno a valores bajos del eje Y, lo que indica que la segunda componente principal tiene poca varianza.
Evitar confiar únicamente en PCA: Aunque PCA es útil para visualización, no es ideal para tareas de clasificación, ya que no optimiza la separación entre clases. Mantén todas las variables originales en experimentos posteriores.
Explorar modelos no lineales: Dado el solapamiento entre clases, experimenta con modelos que permitan fronteras no lineales, como RandomForestClassifier, SVC con kernel no lineal (rbf), o XGBoost.
```

```python
import warnings
import time
import pandas as pd
import numpy as np

# Sklearn imports
from sklearn.model_selection import StratifiedKFold, cross_validate, RandomizedSearchCV
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import make_scorer, accuracy_score, f1_score, precision_score, recall_score
from sklearn.base import clone

# Clasificadores
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import (
    RandomForestClassifier, ExtraTreesClassifier, AdaBoostClassifier,
    GradientBoostingClassifier
)
from sklearn.neural_network import MLPClassifier
from xgboost import XGBClassifier

# Configuración de warnings
warnings.filterwarnings("ignore", category=UserWarning)
warnings.filterwarnings("ignore", message=".gpu_hist.")
warnings.filterwarnings("ignore", message="Parameters: .not used.")
warnings.filterwarnings("ignore", message="The total space of parameters .* is smaller than .*")
warnings.filterwarnings("ignore", category=FutureWarning)


def create_preprocessor(cat_features, num_features):
    """
    Crea el preprocesador para las características numéricas y categóricas
    """
    # Transformador para características numéricas
    num_transformer = Pipeline([
        ('imputer', SimpleImputer(strategy='median')),
        ('scaler', StandardScaler())
    ])

    # Transformador para características categóricas
    cat_transformer = Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('encoder', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
    ])

    # Combinación de transformadores
    preprocessor = ColumnTransformer([
        ('num', num_transformer, num_features),
        ('cat', cat_transformer, cat_features)
    ], remainder='drop')

    return preprocessor


def get_pipe(clf, cat_features, num_features):
    """
    Crea un pipeline completo para el clasificador dado
    """
    preprocessor = create_preprocessor(cat_features, num_features)

    return Pipeline([
        ("preprocessor", preprocessor),
        ("clf", clf)
    ])


def setup_metrics_and_cv():
    """
    Configura las métricas de evaluación y la validación cruzada
    """
    scoring = {
        "f1": make_scorer(f1_score, zero_division=0),
        "accuracy": make_scorer(accuracy_score),
        "recall": make_scorer(recall_score, zero_division=0),
        "precision": make_scorer(precision_score, zero_division=0)
    }

    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

    return scoring, cv


def create_classifiers():
    """
    Crea una lista de clasificadores para evaluar
    """
    classifiers = [
        LogisticRegression(random_state=42, max_iter=2000, class_weight='balanced'),
        KNeighborsClassifier(),
        DecisionTreeClassifier(random_state=42, class_weight='balanced'),
        RandomForestClassifier(random_state=42, n_jobs=-1, class_weight='balanced'),
        ExtraTreesClassifier(random_state=42, n_jobs=-1, class_weight='balanced'),
        AdaBoostClassifier(random_state=42),
        GradientBoostingClassifier(random_state=42),
        MLPClassifier(random_state=42, max_iter=1000),
        XGBClassifier(random_state=42, n_jobs=-1, eval_metric='logloss')
    ]

    return classifiers


def get_param_grids():
    """
    Define los espacios de búsqueda para cada clasificador
    """
    param_grids = {
        "LogisticRegression": [
            {"clf__penalty": ["l1"], "clf__solver": ["liblinear", "saga"],
             "clf__C": np.logspace(-3, 3, 10), "clf__class_weight": ["balanced"],
             "clf__max_iter": [2000], "clf__random_state": [42]},
            {"clf__penalty": ["l2"], "clf__solver": ["liblinear", "lbfgs", "newton-cg", "saga"],
             "clf__C": np.logspace(-3, 3, 10), "clf__class_weight": ["balanced"],
             "clf__max_iter": [2000], "clf__random_state": [42]},
            {"clf__penalty": [None], "clf__solver": ["lbfgs", "newton-cg", "saga"],
             "clf__class_weight": ["balanced"],
             "clf__max_iter": [2000], "clf__random_state": [42]}
        ],
        "KNeighborsClassifier": {
            "clf__n_neighbors": [3, 5, 7, 9, 11],
            "clf__weights": ["uniform", "distance"],
            "clf__p": [1, 2],
            "clf__metric": ["minkowski", "euclidean", "manhattan"]
        },
        "DecisionTreeClassifier": {
            "clf__max_depth": [None, 5, 10, 15, 20],
            "clf__min_samples_split": [2, 5, 10],
            "clf__min_samples_leaf": [1, 2, 4, 8],
            "clf__criterion": ["gini", "entropy", "log_loss"],
            "clf__class_weight": ["balanced"],
            "clf__random_state": [42]
        },
        "RandomForestClassifier": {
            "clf__n_estimators": [100, 200, 300],
            "clf__max_depth": [None, 10, 20, 30],
            "clf__min_samples_split": [2, 5, 10],
            "clf__min_samples_leaf": [1, 2, 4],
            "clf__class_weight": ["balanced", "balanced_subsample"],
            "clf__random_state": [42]
        },
        "ExtraTreesClassifier": {
            "clf__n_estimators": [100, 200, 300],
            "clf__max_depth": [None, 10, 20, 30],
            "clf__min_samples_split": [2, 5, 10],
            "clf__min_samples_leaf": [1, 2, 4],
            "clf__class_weight": ["balanced", "balanced_subsample"],
            "clf__random_state": [42]
        },
        "AdaBoostClassifier": {
            "clf__n_estimators": [50, 100, 200],
            "clf__learning_rate": [0.01, 0.1, 0.5, 1.0],
            "clf__random_state": [42]
        },
        "GradientBoostingClassifier": {
            "clf__n_estimators": [100, 200, 300],
            "clf__learning_rate": [0.01, 0.05, 0.1],
            "clf__max_depth": [3, 5, 7],
            "clf__subsample": [0.7, 0.8, 0.9, 1.0],
            "clf__random_state": [42]
        },
        "MLPClassifier": {
            "clf__hidden_layer_sizes": [(50,), (100,), (50, 25), (100, 50)],
            "clf__activation": ["relu", "tanh"],
            "clf__solver": ["adam"],
            "clf__learning_rate_init": [0.001, 0.01],
            "clf__alpha": [0.0001, 0.001, 0.01],
            "clf__max_iter": [1000],
            "clf__random_state": [42]
        },
        "XGBClassifier": {
            "clf__n_estimators": [100, 200, 300],
            "clf__max_depth": [3, 5, 7],
            "clf__learning_rate": [0.01, 0.05, 0.1],
            "clf__subsample": [0.7, 0.8, 0.9, 1.0],
            "clf__colsample_bytree": [0.7, 0.8, 0.9, 1.0],
            "clf__reg_alpha": [0, 0.1, 0.5],
            "clf__reg_lambda": [0, 0.1, 0.5],
            "clf__random_state": [42]
        }
    }

    return param_grids


def evaluate_models(X, y, cat_features, num_features, n_iter=10):
    """
    Evalúa múltiples clasificadores y devuelve los resultados
    """
    # Configuración
    scoring, cv = setup_metrics_and_cv()
    classifiers = create_classifiers()
    param_grids = get_param_grids()
    results = {}
    best_models = {}

    print(f"\n{'=' * 60}")
    print(f"🚀 EVALUACIÓN DE MODELOS BASELINE")
    print(f"{'=' * 60}")

    # Loop para cada clasificador
    for clf in classifiers:
        name = clf.__class__.__name__
        print(f"\n{'-' * 60}")
        print(f"🔍 Modelo: {name}")

        # Crear pipeline
        pipe = get_pipe(clf, cat_features, num_features)
        best_params = None

        # RandomizedSearchCV si hay grid definido
        if name in param_grids:
            print("⚙️  Running RandomizedSearchCV...")
            search = RandomizedSearchCV(
                estimator=pipe,
                param_distributions=param_grids[name],
                n_iter=n_iter,
                scoring=make_scorer(f1_score, zero_division=0),
                cv=cv,
                n_jobs=-1,
                random_state=42,
                verbose=1
            )
            start = time.time()
            search.fit(X, y)
            elapsed = time.time() - start

            best_params = search.best_params_
            final_pipe = search.best_estimator_
            print(f"✅ Best F1: {search.best_score_:.4f}  (time: {elapsed:.1f}s)")
            print("🏁 Mejores parámetros:")
            for param, value in best_params.items():
                print(f"   - {param}: {value}")
        else:
            print("ℹ️  Sin tuning, usando parámetros por defecto.")
            final_pipe = clone(pipe)
            final_pipe.fit(X, y)  # Entrenar el modelo con parámetros por defecto

        # Guardar el mejor modelo
        best_models[name] = final_pipe

        # Evaluación final con validación cruzada
        print("📊 Evaluando modelo final...")
        cv_res = cross_validate(
            final_pipe, X, y,
            cv=cv, scoring=scoring, return_train_score=False, n_jobs=-1
        )

        # Guardar resultados
        results[name] = {
            "best_params": best_params,
            "f1_mean": np.mean(cv_res["test_f1"]),
            "f1_std": np.std(cv_res["test_f1"]),
            "accuracy_mean": np.mean(cv_res["test_accuracy"]),
            "accuracy_std": np.std(cv_res["test_accuracy"]),
            "recall_mean": np.mean(cv_res["test_recall"]),
            "recall_std": np.std(cv_res["test_recall"]),
            "precision_mean": np.mean(cv_res["test_precision"]),
            "precision_std": np.std(cv_res["test_precision"])
        }

    # Mostrar ranking final
    print(f"\n{'=' * 60}")
    print("📊 RANKING FINAL TRAS TUNING:\n")
    print(f"{'Modelo':>25} | {'F1':^15} | {'Accuracy':^15} | {'Recall':^15} | {'Precision':^15}")
    print(f"{'-' * 25:>25}-+-{'-' * 15:^15}-+-{'-' * 15:^15}-+-{'-' * 15:^15}-+-{'-' * 15:^15}")

    # Ordenar modelos por F1 Score
    sorted_results = sorted(results.items(), key=lambda x: x[1]["f1_mean"], reverse=True)

    for name, stats in sorted_results:
        print(f"{name:>25} | "
              f"{stats['f1_mean']:.3f} ± {stats['f1_std']:.3f} | "
              f"{stats['accuracy_mean']:.3f} ± {stats['accuracy_std']:.3f} | "
              f"{stats['recall_mean']:.3f} ± {stats['recall_std']:.3f} | "
              f"{stats['precision_mean']:.3f} ± {stats['precision_std']:.3f}")

    # Obtener el mejor modelo según F1
    best_model_name = sorted_results[0][0]
    best_model = best_models[best_model_name]

    print(f"\n✨ El mejor modelo es: {best_model_name} con F1: {results[best_model_name]['f1_mean']:.4f}")

    return results, best_models, best_model_name, best_model


# Ejemplo de uso (para ser completado con tu código de carga de datos)
def main():
    # Cargar los datos
    train_df = pd.read_csv('train.csv')

    # Crear la variable objetivo (target) - Si el precio es > 20000 JOD
    train_df['Target'] = (train_df['Price'] > 20000).astype(int)

    # Definir características excluyendo columnas no deseadas
    exclude_cols = ['Price', 'Target']
    if '' in train_df.columns:
        exclude_cols.append('')

    # Identificar características por tipo
    num_features = train_df.select_dtypes(include=['int64', 'float64']).columns.tolist()
    cat_features = train_df.select_dtypes(include=['object']).columns.tolist()

    # Filtrar columnas excluidas
    num_features = [col for col in num_features if col not in exclude_cols]
    cat_features = [col for col in cat_features if col not in exclude_cols]

    # Preparar datos para el modelado
    X = train_df.drop(exclude_cols, axis=1)
    y = train_df['Target']

    # Evaluar modelos
    results, best_models, best_model_name, best_model = evaluate_models(
        X, y, cat_features, num_features, n_iter=10
    )

    return results, best_models, best_model_name, best_model


if __name__ == "__main__":
    results, best_models, best_model_name, best_model = main()
```
